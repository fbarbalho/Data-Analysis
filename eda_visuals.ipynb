```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Carregue o conjunto de dados
wine_df = pd.read_csv('winequality_edited.csv',sep=';')

wine_df.head()
```

### Diagrama de dispersão da qualidade associada a diversas características

```python
# Gráfico de dispersão, relacionando qualidade com PH
```

```python
plt.scatter(x = wine_df['quality'], y = wine_df['pH'],alpha=0.5)
plt.xlabel("quality")
plt.ylabel("pH")
plt.show()
```

```python
plt.scatter(x = wine_df['quality'], y = wine_df['volatile_acidity'],alpha=0.5)
plt.xlabel("quality")
plt.ylabel("volatile_acidity")
plt.show()
```

```python
plt.scatter(x = wine_df['quality'], y = wine_df['residual_sugar'],alpha=0.5)
plt.xlabel("quality")
plt.ylabel("residual_sugar")
plt.show()
```

```python
plt.scatter(x = wine_df['quality'], y = wine_df['alcohol'],alpha=0.5)
plt.xlabel("quality")
plt.ylabel("alcohol")
plt.show()
```
